```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import collections
import csv
with open('dataset/amazon_jobs.csv',encoding = 'UTF-8') as file_obj:
    file_data = csv.DictReader(file_obj,skipinitialspace = True)
    d ={}
    for row in file_data:
        city = row['location'].split(',')
        if city[0]=='IN':
            key = row['location'].split(',')[2]
            if key in d:
                d[key]+=1
            else:
                d[key]=1
    
            
                
        
    oo = collections.OrderedDict(sorted(d.items()))
    plt.pie(oo.values(),labels=d.keys(),autopct = '%0.2f')
    plt.show()
```
